```python
seps = ['\n'*x for x in reversed(range(1,2))]+[' ','']
```

```python
txt_splitter = RecursiveCharacterTextSplitter(separators=seps,
                                              chunk_size = 1000, 
                                              chunk_overlap = 200,
                                             length_function = len)
```

```python
'a\n'*10
```

```python
txt["content"][11360:12000]
```

```python
chunks = txt_splitter.create_documents([txt["content"][11360:]])
```

```python
chunks[2]
```

```python
embeddings_model = HuggingFaceInstructEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2',
                                                  model_kwargs={'device':'cpu'},
                                                encode_kwargs={'normalize_embeddings':False})
```

```python
chunks_txt = [chunk.page_content for chunk in chunks]
```

```python
embeddings = embeddings_model.embed_documents(chunks_txt)
```

```python
len(embeddings)
```

```python
chromadb = Chroma.from_documents(chunks, embeddings_model)
```
